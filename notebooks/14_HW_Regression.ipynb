Построим модель Лассо, для определения наиболее важных признаков:

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_squared_error

RANDOM_SEED = 42  # Задаем фиксированное значение для воспроизводимости

def Lasso_alpha_selection(data):
    # Разделение данных на обучающую и тестовую выборки
    data_train, data_test = train_test_split(data, test_size=0.3, random_state=RANDOM_SEED, shuffle=True)
    
    # Извлечение целевой переменной
    y_train = data_train.pop("MEDV")
    x_train = data_train

    y_test = data_test.pop("MEDV")
    x_test = data_test

    # Определение диапазона значений alpha для подбора
    alpha_values = np.logspace(-4, 1, 50)  # От 0.0001 до 10, 50 значений

    # Настройка модели Lasso для GridSearchCV
    lasso = Lasso()
    param_grid = {'alpha': alpha_values}

    # Использование GridSearchCV для подбора альфа
    grid_search = GridSearchCV(estimator=lasso, param_grid=param_grid, scoring='r2', cv=5)
    grid_search.fit(x_train, y_train)

    # Лучший параметр alpha
    best_alpha = grid_search.best_params_['alpha']
    print(f'Лучший alpha: {best_alpha}')

    # Обучение модели с лучшим alpha
    best_lasso_model = Lasso(alpha=best_alpha)
    best_lasso_model.fit(x_train, y_train)

    # Предсказание на тестовой выборке
    y_pred = best_lasso_model.predict(x_test)

    # Оценка модели
    r2 = r2_score(y_test, y_pred)
    print(f'r2: {r2}')
    mae = mean_absolute_error(y_test,y_pred)
    # Вывод коэффициентов
    coefficients = pd.Series(best_lasso_model.coef_, index=x_train.columns)
    print("Коэффициенты Lasso с лучшим alpha:")
    print(coefficients)

    significant_coefficients = coefficients[abs(coefficients) > 0.1]
    print("\nЗначимые коэффициенты:")
    print(significant_coefficients)

# Предполагается, что df_data_RMTAX уже загружен и содержит необходимые данные
Lasso_alpha_selection(df_data_new)
```

## Предобработка

Напишу класс предобработки
В него я добавлю приведение к нормальному Гауссовскому распределению. 
def normalize_gaussian
Суть ее в том, что если в признаке содержатся отрицательные значения - происходит преобразование Yeo-Johnson, если положительные Z-преобразование
Также добавил LSTAT_poly_2 которые является дополнительным признаком второй степени полинома LSTAT

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PowerTransformer
import pandas as pd


class DataPreprocessing:
    def __init__(self):
        self.scaler = MinMaxScaler()
        self.poly_features = PolynomialFeatures(degree=2, include_bias=False)
        self.min_val = None
        self.max_val = None
        self.standard_scaler = StandardScaler()  # Для Z-преобразования
        self.power_transformer = PowerTransformer(method='yeo-johnson')  # Для Yeo-Johnson
    
    def fit(self, df) -> None:
        # Скопируем исходные данные, чтобы не изменять их
        df_copy = df.copy()
        
        # Логарифмируем признак DIS и создаем новый признак DIS_log
        df_copy[['DIS_GAUSS', 'NOX_GAUSS', 'LSTAT_GAUSS']] = self.normalize_gaussian(df_copy[['DIS', 'NOX', 'LSTAT']])
        # Генерируем полиномиальный признак для LSTAT
        LSTAT_poly = self.poly_features.fit_transform(df_copy[['LSTAT']])
        df_copy['LSTAT_poly_2'] = LSTAT_poly[:, 1]  # Второй столбец - LSTAT^2

        # Масштабируем все признаки
        self.scaler.fit(df_copy)
        
        # Сохраняем минимум и максимум для использования в transform
        self.min_val = self.scaler.data_min_
        self.max_val = self.scaler.data_max_

    def transform(self, df) -> pd.DataFrame:
        # Скопируем исходные данные, чтобы не изменять их
        df_copy = df.copy()
        
        # Логарифмируем признак DIS и создаем новый признак DIS_log
        df_copy[['DIS_GAUSS', 'NOX_GAUSS', 'LSTAT_GAUSS']] = self.normalize_gaussian(df_copy[['DIS', 'NOX', 'LSTAT']])
        # Генерируем полиномиальный признак для LSTAT
        LSTAT_poly = self.poly_features.transform(df_copy[['LSTAT']])
        df_copy['LSTAT_poly_2'] = LSTAT_poly[:, 1]  # Второй столбец - LSTAT^2

        # Масштабируем все признаки
        df_scaled = self.scaler.transform(df_copy)
        df_copy = pd.DataFrame(df_scaled, columns=df_copy.columns, index=df.index)
        return df_copy
    def normalize_gaussian(self, df):
        if (df < 0).any().any():  # Проверка на наличие отрицательных значений
            return self.power_transformer.fit_transform(df)
        else:
            return self.standard_scaler.fit_transform(df)
```
